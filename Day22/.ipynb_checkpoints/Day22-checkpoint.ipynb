```python
# Import the grid!

# Set Grid dimensions
gridWidth = 0
gridHeight = 0
with open("map.txt") as gridMap:
    for line in gridMap:
        gridWidth = max(len(line) - 1, gridWidth)
        gridHeight += 1

grid = [["o" for x in range(gridWidth)] for y in range(gridHeight)]

with open("map.txt") as gridMap:
    for y, line in enumerate(gridMap):
        for x, char in enumerate(line):
            match(char):
                case " ":
                    grid[y][x] = "o"
                case ".":
                    grid[y][x] = "."
                case "#":
                    grid[y][x] = "#"
                    
# Import Commands
commandList = []
with open("Commands.txt") as cmds:
    for line in cmds:
        cmd = ""
        for char in line:
            if char == "R":
                commandList.append(int(cmd))
                commandList.append("R")
                cmd = ""
            elif char == "L":
                commandList.append(int(cmd))
                commandList.append("L")
                cmd = ""
            else:
                cmd += char
    commandList.append(int(cmd))
                
#print(commandList)

startDir = 0
startPos = (-1, -1)
#Set true StartPos
for i, x in enumerate(grid[0]):
    if x == ".":
        startPos = (i, 0)
        break
print(startPos)

# directions: R - 0, D - 1, L - 2, U - 3
currentPos = startPos
currentDir = startDir
for command in commandList:
    #print(currentDir)
    #print(currentPos)
    if (command == "L"):
        currentDir = (currentDir - 1) % 4
    elif (command == "R"):
        currentDir = (currentDir + 1) % 4
    else:
        for n in range(command):
            #try to move
            match(currentDir):
                case 0: # Right
                    target = [(currentPos[0] + 1) % gridWidth, currentPos[1]]
                case 1: # Down
                    target = [currentPos[0], (currentPos[1] + 1) % gridHeight]
                case 2: # Left
                    target = [(currentPos[0] - 1) % gridWidth, currentPos[1]]
                case 3: # Up
                    target = [currentPos[0], (currentPos[1] - 1) % gridHeight]
            match grid[target[1]][target[0]]:
                case "#":
                    continue
                case ".":
                    currentPos = target
                case "o": # Warp to next spot if not a #
                    match(currentDir):
                        case 0:
                            while grid[target[1]][target[0]] == "o":
                                target[0] = (target[0] + 1) % gridWidth
                        case 1:
                            while grid[target[1]][target[0]] == "o":
                                target[1] = (target[1] + 1) % gridHeight
                        case 2:
                            while grid[target[1]][target[0]] == "o":
                                target[0] = (target[0] - 1) % gridWidth
                        case 3:
                            while grid[target[1]][target[0]] == "o":
                                target[1] = (target[1] - 1) % gridHeight
                    if grid[target[1]][target[0]] == "#":
                        break
                    elif grid[target[1]][target[0]] == ".":
                        currentPos = target
                   
#icon = ""                        
#match(currentDir):
#    case 0:
#        icon = ">"
#    case 1:
#        icon = "V"
#    case 2:
#        icon = "<"
#    case 3:
#        icon = "^"
#grid[currentPos[1]][currentPos[0]] = icon
print(currentDir)
print(currentPos)
print(4 * (currentPos[0]+1) + 1000 * (currentPos[1]+1) + currentDir)

#for y in grid:
#    l = ""
#    for x in y:
#        l += x
#    print(l)
```
