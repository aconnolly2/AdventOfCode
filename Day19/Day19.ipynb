```python
class robot:
    def __init__(self, botType):
        self.botType = botType
        
class blueprint:

    def __init__(self, oreBotOreCost, clayBotOreCost, obsBotOreCost, obsBotClayCost, geoBotOreCost, geoBotObsCost):
        self.oreBotOreCost = oreBotOreCost
        self.clayBotOreCost = clayBotOreCost
        self.obsBotOreCost = obsBotOreCost
        self.obsBotClayCost = obsBotClayCost
        self.geoBotOreCost = geoBotOreCost
        self.geoBotObsCost = geoBotObsCost
        self.robots = [robot("ore")]
        self.ore = 0
        self.clay = 0
        self.obs = 0
        self.geo = 0
        self.oreBots = 1
        self.clayBots = 0
        self.obsBots = 0
        self.geoBots = 0
        self.factoryQ = []
        
    def buildRobot(self, botType):
        print(f"Building Robot: {botType}")
        match(botType):
            case "ore":
                if self.ore >= self.oreBotOreCost:
                    self.ore -= self.oreBotOreCost
                    self.factoryQ.append(robot("ore"))
                    self.oreBots += 1
            case "clay":
                if self.ore >= self.clayBotOreCost:
                    self.ore -= self.clayBotOreCost
                    self.factoryQ.append(robot("clay"))
                    self.clayBots += 1
            case "obs":
                if self.ore >= self.obsBotOreCost and self.clay >= self.obsBotClayCost:
                    self.ore -= self.obsBotOreCost
                    self.clay -= self.obsBotClayCost
                    self.factoryQ.append(robot("obs"))
                    self.obsBots += 1
            case "geo":
                if self.ore >= self.geoBotOreCost and self.obs >= self.geoBotObsCost:
                    self.ore -= self.geoBotOreCost
                    self.obs -= self.geoBotObsCost
                    self.factoryQ.append(robot("geo"))
                    self.geoBots += 1
                    
    def collectMinerals(self):
        for bot in self.robots:
            match(bot.botType):
                case "ore":
                    self.ore += 1
                case "clay":
                    self.clay += 1
                case "obs":
                    self.obs += 1
                case "geo":
                    self.geo += 1
                    
    def getOptimalGeos(self, maxTime):
        time = 0
        while time < maxTime:
            #Strategy:
            if self.ore >= self.geoBotOreCost and self.obs >= self.geoBotObsCost:
                self.buildRobot("geo")
            elif self.ore >= self.obsBotOreCost and self.clay >= self.obsBotClayCost and self.obs <= self.geoBotObsCost + self.obsBots:
                self.buildRobot("obs")
            elif self.ore >= self.clayBotOreCost and self.clay <= self.obsBotClayCost + self.clayBots:
                self.buildRobot("clay")
            elif self.ore >= self.oreBotOreCost:
                self.buildRobot("ore")
            
            self.collectMinerals()
            if len(self.factoryQ) > 0:
                self.robots.append(self.factoryQ.pop())
            time += 1
            print(f"{time}: Bots: {len(self.robots)}; Ore: {self.ore}; Clay: {self.clay}; Obsidian: {self.obs}; Geodes: {self.geo}")
        print(f"Ore: {self.ore}; Clay: {self.clay}; Obsidian: {self.obs}; Geodes: {self.geo}")
            

blueprints = []
with open("test19.txt") as bps:
    for line in bps:
        #6: Ore - Ore
        #12: Clay - Ore
        #18: Obs - Ore
        #21: Obs - Clay
        #27: Geo - Ore
        #30: Geo - Obs
        data = line.split(" ")
        oreBotOreCost = int(data[6])
        clayBotOreCost = int(data[12])
        obsBotOreCost = int(data[18])
        obsBotClayCost = int(data[21])
        geoBotOreCost = int(data[27])
        geoBotObsCost = int(data[30])
        bp = blueprint(oreBotOreCost, clayBotOreCost, obsBotOreCost, obsBotClayCost, geoBotOreCost, geoBotObsCost)
        blueprints.append(bp)
        
for i, bp in enumerate(blueprints):
    print(f"Blueprint {i}")
    bp.getOptimalGeos(24)
```
