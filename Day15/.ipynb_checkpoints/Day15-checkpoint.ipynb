```python
import math

def manhattanDistance(a, b):
    return abs(a[0]-b[0])+abs(a[1]-b[1])

sensors = []
beacons = []
with open("Test.txt") as inp:
    for line in inp:
        l = line.split(" ")
        sX = int(l[2].split("=")[1][:-1])
        sY = int(l[3].split("=")[1][:-1])
        sensors.append([sX, sY])
        bX = int(l[8].split("=")[1][:-1])
        bY = int(l[9].split("=")[1].strip())
        beacons.append([bX, bY])

manhattanDistances = []
# Add Sensors and Beacons
for s, b in zip(sensors,beacons):
    manhattanDistances.append(manhattanDistance(s,b))
    
gridMinX = sensors[0][0]-manhattanDistances[0]
gridMinY = sensors[0][1]-manhattanDistances[0]
gridMaxX = sensors[0][0]+manhattanDistances[0]
gridMaxY = sensors[0][1]+manhattanDistances[0]
# Set Grid Dimensions
i = 0
for s, b in zip(sensors,beacons):
    gridMinX = min(s[0]-manhattanDistances[i], gridMinX)
    gridMinY = min(s[1]-manhattanDistances[i], gridMinY)
    gridMaxX = max(s[0]+manhattanDistances[i], gridMaxX)
    gridMaxY = max(s[1]+manhattanDistances[i], gridMaxX)
    i += 1
    #print(f"{s} [{b[0]+s[0]},{b[1]+s[1]}]")

#print(manhattanDistances)
    
gridWidth = gridMaxX - gridMinX
gridHeight = gridMaxY - gridMinY
#print(f"{gridMinX},{gridMinY} x {gridMaxX},{gridMaxY}")
grid = [["." for x in range(gridWidth + 1)] for y in range(gridHeight + 1)]

gridOffsetX = 0-gridMinX
gridOffsetY = 0-gridMinY

for s, b in zip(sensors,beacons):
    grid[s[1]+gridOffsetY][s[0]+gridOffsetX] = "S"
    grid[b[1]+gridOffsetY][b[0]+gridOffsetX] = "B"
    
# Set Sensor Detections
for i, s in enumerate(sensors):
    for x in range(s[0] - manhattanDistances[i] + gridOffsetX, s[0] + manhattanDistances[i] + gridOffsetX + 1):
        for y in range(s[1] - manhattanDistances[i] + gridOffsetY, s[1] + manhattanDistances[i] + gridOffsetY + 1):
            mD = manhattanDistance([s[0] + gridOffsetX, s[1] + gridOffsetY], [x,y])
            #print(f"{sensors[0][0]+gridOffsetX},{sensors[0][1]+gridOffsetY} - {x},{y} : {mD}")
            if (mD <= min(manhattanDistances[i], manhattanDistances[i])) and grid[y][x] == ".":\
                grid[y][x] = "#"
    

testRow = 10
total = 0
for x in grid[testRow+gridOffsetY]:
    if x == "#":
        total += 1
        
print(total)

# Draw Grid
for y in grid:
    l = ""
    for x in y:
        l += x
    print(l)
```
