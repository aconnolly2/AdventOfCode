```python
import math

# Get Grid Dimensions
minGridX = 1000
maxGridX = -1
minGridY = 1000
maxGridY = -1
with open("Cave.txt") as cave:
    for line in cave:
        line = line.strip()
        coords = line.split("->")
        for coord in coords:
            c = coord.split(",")
            minGridX = min(minGridX, int(c[0]))
            maxGridX = max(maxGridX, int(c[0]))
            minGridY = min(minGridY, int(c[1]))
            maxGridY = max(maxGridY, int(c[1]))
            
print("Grid Size:")
print(f"Width: {minGridX} - {maxGridX} -> {maxGridX - minGridX}")
print(f"Height: {minGridY} - {maxGridY} -> {maxGridY - minGridY}")

gridWidth = maxGridX - minGridX
gridHeight = maxGridY
widthOffset = minGridX

# Build a Grid
grid = [["." for x in range(gridWidth)] for y in range(gridHeight)]
print(len(grid))
print(len(grid[0]))

# Sand Source
grid[0][500 - widthOffset - 1] = "+" 

# Populate Rocks
with open("Cave.txt") as cave:
    for line in cave:
        line = line.strip()
        coords = line.split("->")
        prevCoord = coords[0].split(",")
        for coord in coords:
            c = coord.split(",")
            xcoord = int(c[0])
            ycoord = int(c[1])
            if int(c[0]) != int(prevCoord[0]): # Horizontal Line
                if xcoord > int(prevCoord[0]):
                    for x in range(xcoord - int(prevCoord[0]) + 1):
                        grid[ycoord - 1][int(prevCoord[0]) - widthOffset + x - 1] = "#"
                elif int(prevCoord[0]) > xcoord:
                    for x in range(int(prevCoord[0]) - xcoord + 1):
                        grid[ycoord - 1][xcoord - widthOffset + x - 1] = "#"
            elif int(c[1]) != int(prevCoord[1]): # Vertical Line
                if ycoord > int(prevCoord[1]):
                    for y in range(ycoord - int(prevCoord[1])):
                        grid[int(prevCoord[1]) + y - 1][xcoord - widthOffset - 1] = "#"
                elif int(prevCoord[1]) > ycoord:
                    for y in range(int(prevCoord[1]) - ycoord):
                        grid[ycoord + y - 1][xcoord - widthOffset - 1] = "#"
            #else: # Fill this coord just in case
                #grid[ycoord - 1][xcoord - widthOffset - 1] = "#"
            prevCoord = c

# Test
#grid[gridHeight - 1][maxGridX - widthOffset - 1] = "!"
#grid[56-1][506 - widthOffset - 1] = "!"

# Drop Sand
total = 0
sandGrain = [500-widthOffset-1,0]
while (sandGrain[1] < gridHeight - 1):
    sX = int(sandGrain[0])
    sY = int(sandGrain[1])
    # Check immediately below:
    if grid[sY+1][sX] == ".":
        sandGrain = [sX,sY+1]
    elif grid[sY+1][sX-1] == ".":
        sandGrain = [sX-1,sY+1]
    elif grid[sY+1][sX+1] == ".":
        sandGrain = [sX+1,sY+1]
    else:
        total += 1
        grid[sY][sX] = "o"
        sandGrain = [500-widthOffset-1,0]

print(f"Total: {total}")
for y in grid:
    line = ""
    for x in y:
        line += x
    print(line)
```

```python
gridWidth = 1000
gridHeight = 166+3

# Build a Grid
grid = [["." for x in range(gridWidth)] for y in range(gridHeight)]
print(len(grid))
print(len(grid[0]))

# Sand Source
grid[0][500 - 1] = "+" 

# Populate Rocks
with open("Cave.txt") as cave:
    for line in cave:
        line = line.strip()
        coords = line.split("->")
        prevCoord = coords[0].split(",")
        for coord in coords:
            c = coord.split(",")
            xcoord = int(c[0])
            ycoord = int(c[1])
            if int(c[0]) != int(prevCoord[0]): # Horizontal Line
                if xcoord > int(prevCoord[0]):
                    for x in range(xcoord - int(prevCoord[0]) + 1):
                        grid[ycoord][int(prevCoord[0]) + x - 1] = "#"
                elif int(prevCoord[0]) > xcoord:
                    for x in range(int(prevCoord[0]) - xcoord + 1):
                        grid[ycoord][xcoord + x - 1] = "#"
            elif int(c[1]) != int(prevCoord[1]): # Vertical Line
                if ycoord > int(prevCoord[1]):
                    for y in range(ycoord - int(prevCoord[1])):
                        grid[int(prevCoord[1]) + y][xcoord - 1] = "#"
                elif int(prevCoord[1]) > ycoord:
                    for y in range(int(prevCoord[1]) - ycoord):
                        grid[ycoord + y][xcoord - 1] = "#"
            else: # Fill this coord just in case
                grid[ycoord][xcoord - 1] = "#"
            prevCoord = c

# Add Floor
for x in range(len(grid[gridHeight-1])):
    grid[gridHeight-1][x] = "#"

# Test
#grid[130-1][475] = "!"
#grid[130-1][480] = "!"

# Drop Sand
total = 0
sandGrain = [500-1,0]
while (grid[0][500-1] != "o"):
    sX = int(sandGrain[0])
    sY = int(sandGrain[1])
    # Check immediately below:
    if grid[sY+1][sX] == ".":
        sandGrain = [sX,sY+1]
    elif grid[sY+1][sX-1] == ".":
        sandGrain = [sX-1,sY+1]
    elif grid[sY+1][sX+1] == ".":
        sandGrain = [sX+1,sY+1]
    else:
        total += 1
        grid[sY][sX] = "o"
        sandGrain = [500-1,0]

print(f"Total: {total}")
# Verify...
vTotal = 0
for y in grid:
    for x in y:
        if x == "o":
            vTotal += 1
print(f"Verified {vTotal}")
        
for y in grid:
    line = ""
    for x in range(400, 580):
        line += y[x]
    print(line)
    
```
