```python
leftMessages = []
rightMessages = []

def compareItems(lItem, rItem):
    #print(f"{lItem} vs {rItem}")
        
    if type(lItem) == list and type(rItem) == int:
        rItem = [rItem]
        
    if type(lItem) == int and type(rItem) == list:
        lItem = [lItem]
    
    #Easy case first
    if type(lItem) == int and type(rItem) == int:
        if lItem == rItem:
            return 0
        if lItem > rItem:
            return 1
        else:
            return -1
            
    if type(lItem) == list and type(rItem) == list:
        for l, r in zip(lItem, rItem):
            result = compareItems(l, r)
            if result != 0:
                return result
            
        
    return 0

def altCompare(lItem, rItem):
    if type(lItem) is list or type(rItem) is list:
        if type(rItem) is not list: rItem = [rItem]
        if type(lItem) is not list: lItem = [lItem]
        for l, r in zip(lItem, rItem):
            c = altCompare(l, r)
            if c != 0: return c
        return len(lItem) - len(rItem)
    else:
        return lItem-rItem
        
def buildMsg(instr):
    newmsg = []
    stack = []
    slvl = 0
    item = ""
    for i, c in enumerate(instr):
        if c == '[':
            stack.append([])
            slvl += 1
        elif c == ']' and stack:
            try:
                stack[slvl - 1].append(int(item))
                item = ""
            except:
                pass
            msg = stack.pop()
            slvl -=1
            if slvl > 0:
                stack[slvl - 1].append(msg)
            else:
                newmsg = msg
        elif c == ',':
            try:
                stack[slvl - 1].append(int(item))
                item = ""
            except:
                pass
        else:
            item += c
    return newmsg

with open("Messages.txt") as msgs:
    for i, line in enumerate(msgs):
        match i % 3:
            case 0:
                leftMessages.append(line.strip())
            case 1:
                rightMessages.append(line.strip())

allmessages = []
total = 0
for i in range(len(leftMessages)):
    leftmsg = buildMsg(leftMessages[i])
    rightmsg = buildMsg(rightMessages[i])
    allmessages.append(leftmsg)
    allmessages.append(rightmsg)
    result = compareItems(leftmsg, rightmsg)
    #print(result)
    if result < 0:
        #print(i + 1)
        total += i + 1
        
print(total)

#for i in range(len(leftMessages)):
#    print(leftMessages[i])
#    print(rightMessages[i])
#    print()
```

```python
from functools import cmp_to_key
allmessages = []
for i in range(len(leftMessages)):
    leftmsg = buildMsg(leftMessages[i])
    rightmsg = buildMsg(rightMessages[i])
    allmessages.append(leftmsg)
    allmessages.append(rightmsg)
allmessages.append([[2]])
allmessages.append([[6]])
sort = sorted(allmessages, key=cmp_to_key(altCompare))

total = 1
for i, item in enumerate(sort):
    #print(item)
    if item == [[2]] or item == [[6]]:
        total *= (i + 1)
        
print(total)
```
