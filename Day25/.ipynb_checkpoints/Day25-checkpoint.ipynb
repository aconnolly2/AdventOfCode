```python
converted = []

with open("xmas.txt") as codedVals:
    toTranslate = []
    for line in codedVals:
        l = []
        line = line.strip()
        for c in line:
            l.insert(0, c)
        newVal = 0
        for i, n in enumerate(l):
            mul = 5**i
            match(n):
                case("2"):
                    newVal += mul*2
                case("1"):
                    newVal += mul*1
                case("0"):
                    pass
                case("-"):
                    newVal -= mul*1
                case("="):
                    newVal -= mul*2
        converted.append(newVal)
        
    
print(converted)
print(sum(converted))
```
