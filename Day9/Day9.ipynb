```python
commandDir = []
commandDist = []

with open("Input.txt", 'r') as inp:
    for line in inp:
        commands = line.split(' ')
        commandDir.append(commands[0])
        commandDist.append(commands[1].strip())
        
```

```python
# Part 1

headPosX = 0
headPosY = 0
tailPosX = 0
tailPosY = 0
visitedPositions = []
for i, l in enumerate(commandDir):
    for move in range(int(commandDist[i])):
        match commandDir[i]:
            case "U":
                headPosY += 1
            case "D":
                headPosY -= 1      
            case "R":
                headPosX += 1
            case "L":
                headPosX -= 1
        # Tail Chase:
        # Check for diagonal edge case first
        if (abs(headPosX - tailPosX) >= 1 and abs(headPosY - tailPosY) >= 2 or abs(headPosX - tailPosX) >= 2 and abs(headPosY - tailPosY) >= 1):
            if (headPosX - tailPosX >= 1):
                tailPosX += 1
            if (headPosX - tailPosX <= -1):
                tailPosX -= 1
            if (headPosY - tailPosY >= 1):
                tailPosY += 1
            if (headPosY - tailPosY <= -1):
                tailPosY -= 1           
            
        else: # Not diagonal, regular movement ok
            if (headPosX - tailPosX >= 2):
                tailPosX += 1
            if (headPosX - tailPosX <= -2):
                tailPosX -= 1
            if (headPosY - tailPosY >= 2):
                tailPosY += 1
            if (headPosY - tailPosY <= -2):
                tailPosY -= 1
            
        #print(f"Head:{headPosX},{headPosY} - Tail:{tailPosX},{tailPosY}")
        jointTCoord = f"{tailPosX},{tailPosY}"
        visitedPositions.append(jointTCoord)
    #print(f"{commandDir[i]} : {commandDist[i]}")
    
print(len(set(visitedPositions)))
```

```python
# Part 2

headPosX = 0
headPosY = 0
tailCount = 9
tailPosX = [0]*tailCount
tailPosY = [0]*tailCount
visitedPositions = []
for i, l in enumerate(commandDir):
    for move in range(int(commandDist[i])):
        match commandDir[i]:
            case "U":
                headPosY += 1
            case "D":
                headPosY -= 1      
            case "R":
                headPosX += 1
            case "L":
                headPosX -= 1
        # Tail Chase:
        # Check for diagonal edge case first comparing first tail segment
        if (abs(headPosX - tailPosX[0]) >= 1 and abs(headPosY - tailPosY[0]) >= 2 or abs(headPosX - tailPosX[0]) >= 2 and abs(headPosY - tailPosY[0]) >= 1):
            if (headPosX - tailPosX[0] >= 1):
                tailPosX[0] += 1
            if (headPosX - tailPosX[0] <= -1):
                tailPosX[0] -= 1
            if (headPosY - tailPosY[0] >= 1):
                tailPosY[0] += 1
            if (headPosY - tailPosY[0] <= -1):
                tailPosY[0] -= 1           
            
        else: # Not diagonal, regular movement ok
            if (headPosX - tailPosX[0] >= 2):
                tailPosX[0] += 1
            if (headPosX - tailPosX[0] <= -2):
                tailPosX[0] -= 1
            if (headPosY - tailPosY[0] >= 2):
                tailPosY[0] += 1
            if (headPosY - tailPosY[0] <= -2):
                tailPosY[0] -= 1
        
        # Loop remaining tail segments:
        for t in range(1, tailCount):      
            if (abs(tailPosX[t-1] - tailPosX[t]) >= 1 and abs(tailPosY[t-1] - tailPosY[t]) >= 2 or abs(tailPosX[t-1] - tailPosX[t]) >= 2 and abs(tailPosY[t-1] - tailPosY[t]) >= 1):
                if (tailPosX[t-1] - tailPosX[t] >= 1):
                    tailPosX[t] += 1
                if (tailPosX[t-1] - tailPosX[t] <= -1):
                    tailPosX[t] -= 1
                if (tailPosY[t-1] - tailPosY[t] >= 1):
                    tailPosY[t] += 1
                if (tailPosY[t-1] - tailPosY[t] <= -1):
                    tailPosY[t] -= 1           
                
            else: # Not diagonal, regular movement ok
                if (tailPosX[t-1] - tailPosX[t] >= 2):
                    tailPosX[t] += 1
                if (tailPosX[t-1] - tailPosX[t] <= -2):
                    tailPosX[t] -= 1
                if (tailPosY[t-1] - tailPosY[t] >= 2):
                    tailPosY[t] += 1
                if (tailPosY[t-1] - tailPosY[t] <= -2):
                    tailPosY[t] -= 1
            
        #print(f"Head:{headPosX},{headPosY} - Tail:{tailPosX},{tailPosY}")
        jointTCoord = f"{tailPosX[8]},{tailPosY[8]}"
        visitedPositions.append(jointTCoord)
    #print(f"{commandDir[i]} : {commandDist[i]}")
    
print(len(set(visitedPositions)))
```
