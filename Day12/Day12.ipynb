```python
# Our Graph/A* function        
class astar:
    def __init__(self, grid):
        self.grid = grid
        
    def h(self, coord, goal):
        return abs(coord[0] - goal[0]) + abs(coord[1] - goal[1])
    
    def coord_value(self, coord):
        val = self.grid[coord[1]][coord[0]]
        if (val == 'S'):
            return 1
        elif (val == 'E'):
            return 26
        else:
            return ord(val) - 96
        
    def coord_tostring(self, coord):
        return f"{coord[0]},{coord[1]}"
    
    def coord_fromstring(self, coordstr):
        coord = []
        coordstr.split(",")
        coord.append(coordstr[0])
        coord.append(coordstr[1])
        return coord
    
    def get_neighbors(self, coord):
        neighbors = []
        myCoord = self.coord_value(coord)
        # Check Up
        if coord[1] > 0:
            upNeighborCoord = []
            upNeighborCoord.append(coord[0])
            upNeighborCoord.append(coord[1] - 1)
            if self.coord_value(upNeighborCoord) <= myCoord + 1:
                neighbors.append(upNeighborCoord)
        # Check Down
        if coord[1] < len(self.grid) - 1:
            downNeighborCoord = []
            downNeighborCoord.append(coord[0])
            downNeighborCoord.append(coord[1] + 1)
            if self.coord_value(downNeighborCoord) <= myCoord + 1:
                neighbors.append(downNeighborCoord)
        # Check Left
        if coord[0] > 0:
            leftNeighborCoord = []
            leftNeighborCoord.append(coord[0] - 1)
            leftNeighborCoord.append(coord[1])
            if self.coord_value(leftNeighborCoord) <= myCoord + 1:
                neighbors.append(leftNeighborCoord)
        # Check Right
        if coord[0] < len(self.grid[0]) - 1:
            rightNeighborCoord = []
            rightNeighborCoord.append(coord[0] + 1)
            rightNeighborCoord.append(coord[1])
            if self.coord_value(rightNeighborCoord) <= myCoord + 1:
                neighbors.append(rightNeighborCoord)
        return neighbors
    
    def fscore(coord, g):
        return g + self.h(coord)
    
    def search(self, start, end):
        open_list = [start]
        closed_list = []
        
        g = {"-1,-1":999999}
        
        parents = {}
        parents[self.coord_tostring(start)] = start
        #print(parents)
        
        for node in open_list:
            g[self.coord_tostring(node)] = 1
            #print(self.h(node, end))
        
        while len(open_list) > 0:
            q = [-1,-1]
            for n in open_list:
                if g[self.coord_tostring(n)] + self.h(n, end) < g[self.coord_tostring(q)] + self.h(q, end):
                    q = n
            
            if q == [-1, -1]:
                #print('No Path')
                return None
            
            if q == end:
                reconst_path = []
                
                while parents[self.coord_tostring(q)] != q:
                    reconst_path.append(q)
                    q = parents[self.coord_tostring(q)]
                    
                reconst_path.append(start)
                
                reconst_path.reverse()
                #print(open_list)
                #print(f"Path found: {reconst_path}")
                return reconst_path
            
            for neighbor in self.get_neighbors(q):
                if neighbor not in open_list and neighbor not in closed_list:
                    open_list.append(neighbor)
                    parents[self.coord_tostring(neighbor)] = q
                    g[self.coord_tostring(neighbor)] = g[self.coord_tostring(q)] + 1
                    
                else:
                    if g[self.coord_tostring(neighbor)] > g[self.coord_tostring(q)] + 1:
                        g[self.coord_tostring(neighbor)] = g[self.coord_tostring(q)] + 1
                        parents[self.coord_tostring(neighbor)] = q
                        
                        if neighbor in closed_list:
                            closed_list.remove(neighbor)
                            open_list.append(neighbor)
            #print(q)
            #print(open_list)
            open_list.remove(q)
            closed_list.append(q)
            
        #print('No path!')
        return None
    
class BFS:
    def __init__(self, grid):
        self.grid = grid
    
    def coord_value(self, coord):
        val = self.grid[coord[1]][coord[0]]
        if (val == 'S'):
            return 1
        elif (val == 'E'):
            return 27
        else:
            return ord(val) - 96
        
    def coord_tostring(self, coord):
        return f"{coord[0]},{coord[1]}"
    
    def coord_fromstring(self, coordstr):
        coord = []
        coordstr.split(",")
        coord.append(coordstr[0])
        coord.append(coordstr[1])
        return coord
    
    def get_neighbors(self, coord):
        neighbors = []
        myCoord = self.coord_value(coord)
        # Check Up
        if coord[1] > 0:
            upNeighborCoord = []
            upNeighborCoord.append(coord[0])
            upNeighborCoord.append(coord[1] - 1)
            if self.coord_value(upNeighborCoord) <= myCoord + 1:
                neighbors.append(upNeighborCoord)
        # Check Down
        if coord[1] < len(self.grid) - 1:
            downNeighborCoord = []
            downNeighborCoord.append(coord[0])
            downNeighborCoord.append(coord[1] + 1)
            if self.coord_value(downNeighborCoord) <= myCoord + 1:
                neighbors.append(downNeighborCoord)
        # Check Left
        if coord[0] > 0:
            leftNeighborCoord = []
            leftNeighborCoord.append(coord[0] - 1)
            leftNeighborCoord.append(coord[1])
            if self.coord_value(leftNeighborCoord) <= myCoord + 1:
                neighbors.append(leftNeighborCoord)
        # Check Right
        if coord[0] < len(self.grid[0]) - 1:
            rightNeighborCoord = []
            rightNeighborCoord.append(coord[0] + 1)
            rightNeighborCoord.append(coord[1])
            if self.coord_value(rightNeighborCoord) <= myCoord + 1:
                neighbors.append(rightNeighborCoord)
        return neighbors
    
    def search(self, start, end):
        paths=[[start]]
        path_index = 0
        prev_nodes = {self.coord_tostring(start)}
        
        if start == end:
            return paths[0]
        
        while path_index < len(paths):
            current_path = paths[path_index]
            last_node = current_path[-1]
            next_nodes = self.get_neighbors(last_node)
            
            if end in next_nodes:
                current_path.append(end)
                return current_path
            
            for next_node in next_nodes:
                if not self.coord_tostring(next_node) in prev_nodes:
                    new_path = current_path[:]
                    new_path.append(next_node)
                    paths.append(new_path)
                    prev_nodes.add(self.coord_tostring(next_node))
                    
            path_index += 1
            
        return []
    
# Import our data to 2D Array
grid = []
with open("Map.txt") as heightMap:
    for line in heightMap:
        l = []
        line = line.strip()
        for c in line:
            l.append(c)
        grid.append(l)
     
    
start = []
end = []
for y, line in enumerate(grid):
    l = ""
    for x, c in enumerate(line):
        if c == 'S':
            start.append(x)
            start.append(y)
        elif c == 'E':
            end.append(x)
            end.append(y)
        i = f"{ord(c) - 96}"
        l += i
    #print(l)
    
#print(f"{start}: {grid[int(startcoords[1])][int(startcoords[0])]} {end}: {grid[int(endcoords[1])][int(endcoords[0])]}")
#print(start)
graph = astar(grid)
path = graph.search(start, end)
#altgraph = BFS(grid)
#path = altgraph.search(start, end)

for coord in path:
    grid[coord[1]][coord[0]] = grid[coord[1]][coord[0]].upper()
    
for line in grid:
    l = ""
    for c in line:
        if c.isupper():
            c = "\033[92m" + c
        else:
            c = '\033[91m' + c
        l += c
    print(l)
    
print(len(path)-1)
```
