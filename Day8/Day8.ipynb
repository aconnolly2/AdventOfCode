```python
def emptyGrid(g):
    eg = []
    for line in g:
        l = ""
        for char in line:
            l += "0"
        eg.append(list(l))
    return eg

def combineGrids(grids):
    result = emptyGrid(grids[0])
    for grid in grids:
        for i, line in enumerate(grid):
            for j, char in enumerate(line):
                result[i][j] = int(result[i][j]) + int(char)
                
    return result

def countVisible(g):
    total = 0
    for line in g:
        for char in line:
            if char > 0:
                total += 1
    
    return total
    

grid = []
with open('Input.txt', 'r') as inp:
    for line in inp:
        grid.append(list(line.strip()))
        
```

```python
#Part 1

total = 0

#evaluate left
leftGrid = emptyGrid(grid)
for i, line in enumerate(grid):
    biggestTree = -1
    for j, char in enumerate(line):
        if int(char) > int(biggestTree):
            biggestTree = char
            leftGrid[i][j] = 1
            

#evaluate right
rightGrid = emptyGrid(grid)
for i, line in enumerate(grid):
    biggestTree = -1
    for j, char in reversed(list(enumerate(line))):
        if int(char) > int(biggestTree):
            biggestTree = char
            rightGrid[i][j] = 1
            
#evaluate up
upGrid = emptyGrid(grid)
biggestTrees = [-1]*len(grid[0])
for i, line in enumerate(grid):
    for j, char in enumerate(line):
        if int(char) > int(biggestTrees[j]):
            biggestTrees[j] = int(char)
            upGrid[i][j] = 1
            

#evaluate down
downGrid = emptyGrid(grid)
biggestTrees = [-1]*len(grid[0])
for i, line in reversed(list(enumerate(grid))):
    for j, char in enumerate(line):
        if int(char) > int(biggestTrees[j]):
            biggestTrees[j] = int(char)
            downGrid[i][j] = 1

grids = [upGrid, downGrid, leftGrid, rightGrid]

combinedGrid = combineGrids(grids)

print(countVisible(combinedGrid))
```

```python
# Part 2

scenicScores = emptyGrid(grid)

# These ones are just for troubleshooting...
upScores = emptyGrid(grid)
downScores = emptyGrid(grid)
leftScores = emptyGrid(grid)
rightScores = emptyGrid(grid)

for i, line in enumerate(grid):
    for j, char in enumerate(line):
        myTree = int(char)
        #left
        leftScore = 1
        index = j - 1
        if index >= 0:
            leftTree = int(grid[i][index])
        else:
            leftScore = 0
        while leftTree < myTree and index >= 0:
            leftScore += 1
            index = index - 1
            if index >= 0:
                leftTree = int(grid[i][index])
            else:
                leftScore -= 1
        leftScores[i][j] = leftScore
        #right
        rightScore = 1
        index = j + 1
        if index <= len(line) - 1:
            rightTree = int(grid[i][index])
        else:
            rightScore = 0
        while rightTree < myTree and index < len(line):
            rightScore += 1
            index = index + 1
            if index <= len(line) - 1:
                rightTree = int(grid[i][index])
            else:
                rightScore -= 1
        rightScores[i][j] = rightScore
        #up
        upScore = 1
        index = i - 1
        if index >= 0:
            upTree = int(grid[index][j])
        else:
            upScore = 0
        while upTree < myTree and index >= 0:
            upScore += 1
            index = index - 1
            if index >= 0:
                upTree = int(grid[index][j])
            else:
                upScore -= 1
        upScores[i][j] = upScore
        #down
        downScore = 1
        index = i + 1
        if index <= len(grid) - 1:
            downTree = int(grid[index][j])
        else:
            downScore = 0
        while downTree < myTree and index < len(grid):
            downScore += 1
            index = index + 1
            if index <= len(grid) - 1:
                downTree = int(grid[index][j])
            else:
                downScore -= 1
        downScores[i][j] = downScore
        
        score = leftScore * rightScore * upScore * downScore
        scenicScores[i][j] = int(score)
        
highestScore = 0
for line in scenicScores:
    for char in line:
        if int(char) > highestScore:
            highestScore = int(char)
            
print(highestScore)
```
